```python
# Q4) Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(42)
categories = ['A', 'B', 'C']
data = {
    'A': np.random.normal(10, 2, 50),
    'B': np.random.normal(15, 3, 50),
    'C': np.random.normal(20, 4, 50)
}
data_values = [data[cat] for cat in categories]
plt.figure(figsize=(8, 6))
plt.boxplot(data_values)
plt.title('Distribution of Numerical Values Across Categories', fontsize=14)
plt.xlabel('Categories', fontsize=12)
plt.show()
```

```python
# Q5) Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


np.random.seed(42)
n_samples = 100
x1 = np.random.rand(n_samples) * 10
x2 = 2.5 * x1 + np.random.randn(n_samples) * 2
x3 = -1.5 * x1 + np.random.randn(n_samples) * 3
x4 = np.random.rand(n_samples) * 10
data = pd.DataFrame({
    'Feature 1': x1,
    'Feature 2': x2,
    'Feature 3': x3,
    'Feature 4': x4
})
correlation_matrix = data.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix)
plt.title('Correlation Matrix Heatmap', fontsize=14)
plt.show()
```

```python
# Q4) Using the given x and y data, generate a 3D surface plot to visualize the function z= sin(root(x2+y2))
# x = np.linspace(-5, 5, 100)
# y = np.linspace(-5, 5, 100)
# x, y = np.meshgrid(x, y)
# z = np.sin(np.sqrt(x**2 + y**2))
# data = {
#     'X': x.flatten(),
#     'Y': y.flatten(),
#     'Z': z.flatten()
# }
# df = pd.DataFrame(data)

# Ans.
import numpy as np
import pandas as pd
import plotly.graph_objects as go


x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)


z = np.sin(np.sqrt(x**2 + y**2))


data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)


fig = go.Figure(data=[go.Surface(z=z, x=x, y=y, colorscale='viridis')])


fig.update_layout(
    title='3D Surface Plot: z = sin(sqrt(x + y))',
    scene=dict(
        xaxis_title='X-axis',
        yaxis_title='Y-axis',
        zaxis_title='Z-axis'
    )
)


fig.show()
```

# **BOKEH ASSIGNEMENT**

```python
# Q1) Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
# Ans.
from bokeh.plotting import figure, show
from bokeh.io import output_file
import numpy as np


x = np.linspace(0, 10, 100)
y = np.sin(x)

p = figure(title='Sine Wave', x_axis_label='X', y_axis_label='sin(X)', width=700, height=400)


p.line(x, y, legend_label='sin(x)', line_width=2, color='blue')


output_file("sine_wave.html")
show(p)
```

```python
# Q2) Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.
# Ans.
from bokeh.plotting import figure, show
from bokeh.io import output_file
import numpy as np
import pandas as pd
from bokeh.models import ColumnDataSource


np.random.seed(42)
data = {
    'x': np.random.uniform(0, 10, 100),
    'y': np.random.uniform(0, 10, 100),
    'sizes': np.random.randint(5, 30, 100),
    'colors': np.random.randint(0, 256, 100)
}
df = pd.DataFrame(data)


source = ColumnDataSource(df)



p = figure(title='Bokeh Scatter Plot', x_axis_label='X', y_axis_label='Y', width=700, height=400)


p.scatter(
    'x', 'y', source=source,
    size='sizes', color='colors', fill_alpha=0.6
)



show(p)
```

```python
# Q3) Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
# fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
# counts = [20, 25, 30, 35]
# Ans.
from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.models import ColumnDataSource


fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]


source = ColumnDataSource(data={'fruits': fruits, 'counts': counts})

p = figure(x_range=fruits, title='Fruit Counts', x_axis_label='Fruits', y_axis_label='Count', width=700, height=400)


p.vbar(x='fruits', top='counts', width=0.5, source=source, color='blue')



show(p)
```

```python
# Q4) Create a Bokeh histogram to visualize the distribution of the given data.
# data_hist = np.random.randn(1000)
# hist, edges = np.histogram(data_hist, bins=30)

# Ans.
from bokeh.plotting import figure, show
from bokeh.io import output_file
import numpy as np


np.random.seed(42)
data_hist = np.random.randn(1000)


hist, edges = np.histogram(data_hist, bins=30)


p = figure(title='Bokeh Histogram', x_axis_label='Value', y_axis_label='Frequency', width=700, height=400)


p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='blue', line_color='black', alpha=0.6)


show(p)
```
